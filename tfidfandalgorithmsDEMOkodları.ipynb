```python
#DEMO KODLARIMIZ
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Eğitim verisi ve modelin tanımlanması
data = pd.read_excel('/content/sample_data/duygular_stemmed.xlsx')
dataDoc = ["" if isinstance(doc, float) and pd.isna(doc) else doc for doc in data['comments']]
dataClass = data['label']

X_train, X_test, y_train, y_test = train_test_split(dataDoc, dataClass, test_size=0.3, random_state=42)

# TF-IDF vektörleştirici
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1,2), max_df=0.8, min_df=1)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# en iyi Modeli eğit
clf = SVC(kernel='linear', C=1.0)
FinalModel = clf.fit(X_train_tfidf, y_train)

# Kelimeleri sayısal vektörlere dönüştüren fonksiyon
def word_toNumeric(inpText):
    # TF-IDF vektörleştirici ile dönüştür
    X = tfidf_vectorizer.transform(inpText)

    # Tahmin yap
    Prediction = FinalModel.predict(X)

    # Sonuçları DataFrame'e dönüştür
    Result = pd.DataFrame({'Yorumlar': inpText, 'DuyguTahmin': Prediction})
    return Result

# Modelimizi test etmek için bazı sample cümleler yazıyoruz...
SampleExample = [
    " Sürprizi görünce ben şok kaldım öyle birden  ",
    "bir akıllı sensin zaten",
    "köpek fobim var benim çok korkarım",
    "Tabi tabi en çok sen biliyorsun"
]

print(word_toNumeric(inpText=SampleExample))
```
